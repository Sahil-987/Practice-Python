Prompt : What are different list methods in Python ?

In Python, lists are a versatile data structure that can hold an ordered collection of items. Lists come with a variety of built-in methods that allow you to manipulate, modify, and work with the data stored within them. Here are some of the commonly used list methods in Python:

1. `append(item)`: Adds an item to the end of the list.
2. `extend(iterable)`: Adds all items from an iterable (e.g., another list, tuple, or string) to the end of the list.
3. `insert(index, item)`: Inserts an item at a specified index within the list.
4. `remove(item)`: Removes the first occurrence of a specific item from the list.
5. `pop(index=-1)`: Removes and returns the item at the specified index. If no index is provided, the last item is removed.
6. `index(item, start=0, end=len(list))`: Returns the index of the first occurrence of the specified item within the specified range.
7. `count(item)`: Returns the number of occurrences of a specific item in the list.
8. `sort(key=None, reverse=False)`: Sorts the items in the list in ascending order. The `key` parameter can be used to provide a custom sorting function, and `reverse` can be set to `True` to sort in descending order.
9. `reverse()`: Reverses the order of the items in the list.
10. `copy()`: Returns a shallow copy of the list.
11. `clear()`: Removes all items from the list, making it empty.
12. `len(list)`: Returns the number of items in the list.
13. `list(iterable)`: Converts an iterable (e.g., a tuple, string, or another list) into a new list.
14. `max(list)`: Returns the maximum item in the list.
15. `min(list)`: Returns the minimum item in the list.

Example usage:

```python
my_list = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]

my_list.append(8)
my_list.extend([7, 0])
my_list.insert(2, 2.5)
my_list.remove(5)
popped_item = my_list.pop(4)
index_of_3 = my_list.index(3)
count_of_1 = my_list.count(1)
my_list.sort()
my_list.reverse()

print(my_list)
```

Remember that these are just a subset of the methods available for working with lists in Python. Each method serves a specific purpose, allowing you to manipulate and manage your data effectively.
